```python
# Remove two algorithms that are cclearly inefficient and do more analysis on the best most sensible ones.

exp_df2 = exp_df[exp_df.Algorithm.str.contains("IDDFS") == False]
exp_df3 = exp_df2[exp_df2.Algorithm.str.contains("BFS") == False]
exp_df3
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
ax = sns.boxplot(x='Algorithm', y='Total States Viewed', data=exp_df3, color='#99c2a2')
ax = sns.swarmplot(x="Algorithm", y="Total States Viewed", data=exp_df3, color='#7d0013')
plt.show()
```

```python
#this has to be done for each inversion x algorithm combo but code works,

# check normality assumption

import numpy as np 
import scipy.stats as stats
import matplotlib.pyplot as plt

unique_alg = exp_df3['Algorithm'].unique()
for alg in unique_alg:
    stats.probplot(exp_df3[exp_df3['Algorithm'] == alg]['Total States Viewed'], dist="norm", plot=plt)
    plt.title("Probability Plot - " +  alg)
    plt.show()
```

```python
#do for each inversion level

# get ANOVA table as R like output
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Ordinary Least Squares (OLS) model
model = ols('exp_df3["Total States Viewed"] ~ C(Algorithm)', data=exp_df3).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```
